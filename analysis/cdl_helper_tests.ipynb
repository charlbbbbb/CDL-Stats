# Small non-comprehensive tests for functions created in cdl_helpter.py

```python
from cdl_helper import compare_stats, read_in_list

%matplotlib inline
```

```python
import pandas as pd
matchIDs = pd.read_json('../major_ids.json')

matchIDs
```

```python
m1_qual = read_in_list(matchIDs['major1'][0])
m2_qual = read_in_list(matchIDs['major2'][0])
m2_event = read_in_list(matchIDs['major2'][1])
m3_qual = read_in_list(matchIDs['major3'][0])

m1_qual['event'] = ['M1Qual' for i in m1_qual[m1_qual.columns[0]]]
m2_qual['event'] = ['M2Qual' for i in m2_qual[m2_qual.columns[0]]]
m3_qual['event'] = ['M3Qual' for i in m3_qual[m3_qual.columns[0]]]
m2_event['event'] = ['M2Event' for i in m2_event[m2_event.columns[0]]]

m1_qual['setting'] = ['online' for i in m1_qual[m1_qual.columns[0]]]
m2_qual['setting'] = ['online' for i in m2_qual[m2_qual.columns[0]]]
m3_qual['setting'] = ['online' for i in m3_qual[m3_qual.columns[0]]]
m2_event['setting'] = ['lan' for i in m2_event[m2_event.columns[0]]]

complete_df = pd.concat([m1_qual, m2_event, m2_qual, m3_qual])
complete_df.sample(5)
```

```python
atl_pal = {'ATL': 'red'}
fig = compare_stats(complete_df, 'averageSpeed', 'totalDistanceTraveled', legend=True, hidden_spines=['right', 'top'], team='ATL', gamemode='CDL Hardpoint',
                     map='Breenbergh Hotel', sens=1, pal=atl_pal, show_event=True, size=(18, 10))
```

```python
from cdl_helper import CDL_PALETTE
complete_df['kd'] = complete_df['totalKills']/complete_df['totalDeaths']
fig = compare_stats(complete_df, 'kd', 'hillTime', legend=True, hidden_spines=['right', 'top'], sens=0.8, size=(18, 10), gamemode='CDL Hardpoint', pal=CDL_PALETTE)
```

```python
complete_df['kd'] = complete_df['totalKills']/complete_df['totalDeaths']
fig = compare_stats(complete_df, 'kd', 'totalDamageDealt', legend=True, hidden_spines=['right', 'top'], sens=0.6, size=(18, 10), gamemode='CDL Control', pal=CDL_PALETTE)
```
